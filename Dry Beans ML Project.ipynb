<font size="6"><font color="blue"><b>Data Description</font></b></p>

1.   Area (A): The area of a bean zone and the number of pixels within its boundaries.
2.  Perimeter (P): Bean circumference is defined as the length of its border.
3.  Major axis length (L): The distance between the ends of the longest line that can be drawn from a bean.
4.  Minor axis length (l): The longest line that can be drawn from the bean while standing perpendicular to the main axis.
5.  Aspect ratio (K): Defines the relationship between L and l.
6.  Eccentricity (Ec): Eccentricity of the ellipse having the same moments as the region.
7.  Convex area (C): Number of pixels in the smallest convex polygon that can contain the area of a bean seed.
8.  Equivalent diameter (Ed): The diameter of a circle having the same area as a bean seed area.
9.  Extent (Ex): The ratio of the pixels in the bounding box to the bean area.
10. Solidity (S): Also known as convexity. The ratio of the pixels in the convex shell to those found in beans.
11. Roundness (R): Calculated with the following formula: (4piA)/(P^2)
12. Compactness (CO): Measures the roundness of an object: Ed/L
13. ShapeFactor1 (SF1)
14. ShapeFactor2 (SF2)
15. ShapeFactor3 (SF3)
16. ShapeFactor4 (SF4)
17. Class (Seker, Barbunya, Bombay, Cali, Dermosan, Horoz and Sira)

### Handling Outliers using IQR Method

```python
def remove_outliers(data, columns):
    data_clean = data.copy()  # Make a copy of the original DataFrame
    for col in columns:
        # Calculate Q1, Q3, and IQR
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1

        # Calculate lower and upper whiskers
        lower_whisker = Q1 - 1.5 * IQR
        upper_whisker = Q3 + 1.5 * IQR

        # Filter rows where values are within whiskers
        data_clean = data_clean[(data_clean[col] >= lower_whisker) & (data_clean[col] <= upper_whisker)]
    return data_clean
```

```python
dataf = remove_outliers(data, ["Area", "Perimeter", "MajorAxisLength", "MinorAxisLength","AspectRation", "Eccentricity", "ConvexArea","EquivDiameter","Extent","Solidity","roundness","ShapeFactor1","ShapeFactor3","ShapeFactor4"])
```

```python
def cap_outliers(data, columns):
    dataf_cap = dataf.copy()  # Make a copy of the original DataFrame
    for col in columns:
        # Calculate Q1, Q3, and IQR
        Q1 = dataf[col].quantile(0.25)
        Q3 = dataf[col].quantile(0.75)
        IQR = Q3 - Q1

        # Calculate lower and upper whiskers
        lower_whisker = Q1 - 1.5 * IQR
        upper_whisker = Q3 + 1.5 * IQR

        # Cap the outliers to whisker limits
        dataf_cap[col] = dataf[col].clip(lower=lower_whisker, upper=upper_whisker)
    return dataf_cap

columns_to_cap = ["Area", "Perimeter", "MajorAxisLength", "MinorAxisLength","AspectRation", "Eccentricity", "ConvexArea","EquivDiameter","Extent","Solidity","roundness","ShapeFactor4"]
dataf_cap = cap_outliers(data, columns_to_cap)
```

### Handling Class Imbalance with SMOTETomek 

### Why Use SMOTETomek for the Dry Beans Dataset?

-  The Dry Beans dataset consists of various bean types with different characteristics, often leading to class imbalance—some bean types may be underrepresented compared to others. Using SMOTETomek in this dataset helps improve classification performance.
-  SMOTETomek combines oversampling (SMOTE) and undersampling (Tomek Links) to:
  
✅ Generate synthetic examples for underrepresented bean types (SMOTE).
  
✅ Remove redundant majority class samples that are close to minority class samples (Tomek Links).

✅ Improve class separation, reducing overlapping and making classification more accurate.

✅ Prevent model bias, ensuring fair learning across all bean types.

```python

from imblearn.combine import SMOTETomek
X = X_scaled_data  # Features excluding the target
# Apply SMOTETomek to balance the dataset
smote = SMOTETomek(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check the class distribution after oversampling
print("Class distribution after oversampling:")
print(pd.Series(y_resampled).value_counts())
```

```python
# Step 4: Convert the resampled data back to a DataFrame if needed
X_resampled= pd.DataFrame(X_resampled, columns=X.columns)
y_resampled = pd.DataFrame(y_resampled, columns=['Class'])

# Step 5: Optionally, combine the resampled features and target back into a single DataFrame
data_resampled = pd.concat([X_resampled, y_resampled], axis=1)

# Step 6: Check the new class distribution to ensure balance
print(data_resampled['Class'].value_counts())
```

### Insights After Applying SMOTETomek on the Dry Beans Dataset

1. Balanced Class Distribution
   
The dataset has almost equal representation across all classes.

While not perfectly equal, the variation is minimal (3,268 to 3,443), which helps avoid overfitting.

This ensures that the model can learn patterns for all bean types instead of being biased toward the majority class.

### Why the Counts Are Slightly Different?

- SMOTE (oversampling) increases the count of minority classes by generating synthetic data.

- Tomek Links (undersampling) removes ambiguous majority class samples that are too close to minority class samples.
    
- This natural variation (instead of perfect equality) reduces noise and enhances decision boundaries.

### Expected Impact on Model Performance

- Improved Classification Performance
- Higher Recall & Precision for Minority Classes
- Reduced Overfitting Risks

### Performing Hyperparameter tuning for Random Forest Algorithm

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
# Define hyperparameter search space
param_dist = {
    "n_estimators": randint(50, 300),  # Number of trees in the forest
    "max_depth": randint(5, 50),  # Depth of trees
    "min_samples_split": randint(2, 10),  # Min samples required to split a node
    "min_samples_leaf": randint(1, 10),  # Min samples required in a leaf
    "max_features": ["auto", "sqrt", "log2"],  # Number of features considered for best split
    "bootstrap": [True, False]  # Whether bootstrap samples are used when building trees
}

# Define the model
rf_model = RandomForestClassifier(random_state=42)

# Perform Random Search
random_search = RandomizedSearchCV(
    rf_model, 
    param_dist, 
    n_iter=20,  # Number of random configurations to test
    cv=5,  # 5-fold cross-validation
    scoring="accuracy",  # Optimization metric
    n_jobs=-1,  # Use all available cores
    random_state=42
)

# Fit the model
random_search.fit(X_train, y_train)

# Print best parameters and accuracy
print("Best Parameters:", random_search.best_params_)
print("Best Accuracy:", random_search.best_score_)

# Evaluate on test set
best_rf = random_search.best_estimator_
test_accuracy = best_rf.score(X_test, y_test)
print("Test Accuracy:", test_accuracy)
```
